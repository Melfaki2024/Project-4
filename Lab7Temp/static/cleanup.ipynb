```python
import pandas as pd
```

```python
capital_clean = pd.read_csv("CapitalBudgets.csv")
capital_clean = capital_clean.set_index("Categories")
capital_clean
```

```python
for i in range(15):
    capital_clean[capital_clean.columns[i]] = capital_clean[capital_clean.columns[i]].str.replace(',', '')
    capital_clean[capital_clean.columns[i]] = pd.to_numeric(capital_clean[capital_clean.columns[i]])
```

```python
capital_clean
```

```python
capital_clean.to_csv("capital_clean.csv")
```

```python
stormwater_clean = pd.read_csv('Stormwater.csv')
stormwater_clean = stormwater_clean.set_index("Categories")
stormwater_clean
```

```python
stormwater_clean.columns
```

```python
for i in range(15):
    column_name = stormwater_clean.columns[i]
    if stormwater_clean[column_name].dtype == 'object':
        stormwater_clean[column_name] = stormwater_clean[column_name].str.replace(',', '')
        stormwater_clean[column_name] = pd.to_numeric(stormwater_clean[column_name])
```

```python
stormwater_clean.to_csv("stormwater_clean.csv")
```
